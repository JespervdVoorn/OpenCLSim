```python
import datetime, time
import simpy

import shapely.geometry
from simplekml import Kml, Style

import pandas as pd
import openclsim.core as core
import openclsim.model as model

import datetime

import numpy as np

from plot import vessel_planning
```

```python
simulation_start = 0
my_env = simpy.Environment(initial_time=simulation_start)
registry = {}
keep_resources = {}
```

```python
Site = type(
    "Site",
    (
        core.Identifiable,
        core.Log,
        core.Locatable,
        core.HasContainer,
        core.HasResource,
    ),
    {},
)


TransportProcessingResource = type(
    "TransportProcessingResource",
    (
        core.Identifiable,
        core.Log,
        core.ContainerDependentMovable,
        core.Processor,
        core.HasResource,
    ),
    {},
)
```

```python
location_from_site = shapely.geometry.Point(5.1, 52)
location_to_site = shapely.geometry.Point(5, 52.1)

data_from_site = {
    "env": my_env,
    "name": "Winlocatie",
    "geometry": location_from_site,
    "capacity": 20,
    "level": 20,
}

data_from_site_2 = {
    "env": my_env,
    "name": "Winlocatie_2",
    "geometry": location_from_site,
    "capacity": 20,
    "level": 20,
}

data_from_site_3 = {
    "env": my_env,
    "name": "Winlocatie_3",
    "geometry": location_from_site,
    "capacity": 20,
    "level": 20,
}

data_to_site = {
    "env": my_env,
    "name": "Dumplocatie",
    "geometry": location_to_site,
    "capacity": 1000,
    "level": 0,
}


from_site = Site(**data_from_site)
from_site_2 = Site(**data_from_site_2)
from_site_3 = Site(**data_from_site_3)
to_site = Site(**data_to_site)
```

```python
data_hopper = {
    "env": my_env,
    "name": "Hopper 01",
    "geometry": location_from_site,
    "capacity": 4,
    "compute_v": lambda x: 10
}

hopper = TransportProcessingResource(**data_hopper)

data_hopper2 = {
    "env": my_env,
    "name": "Hopper 02",
    "geometry": location_from_site,
    "capacity": 4,
    "compute_v": lambda x: 10
}

hopper2 = TransportProcessingResource(**data_hopper2)
```

```python
requested_resources1 = {}


single_run = [
    model.MoveActivity(**{
        "env": my_env,
        "name": "sailing empty",
        "registry": registry,
        "mover": hopper,
        "destination": from_site,
        "postpone_start": True,
    }),
    model.ShiftAmountActivity(**{
        "env": my_env,
        "name": "Transfer MP",
        "registry": registry,
        "processor": hopper,
        "origin": from_site,
        "destination": hopper,
        "amount": 4,
        "duration": 3600,
        "postpone_start": True,
        "requested_resources": {},
    }),
    model.MoveActivity(**{
        "env": my_env,
        "name": "sailing filled",
        "registry": registry,
        "mover": hopper,
        "destination": to_site,
        "postpone_start": True,
    }),
    model.ShiftAmountActivity(**{
        "env": my_env,
        "name": "Transfer TP",
        "registry": registry,
        "processor": hopper,
        "origin": hopper,
        "destination": to_site,
        "amount": 4,
        "duration": 3600,
        "postpone_start": True,
        "requested_resources": {},
    })
]


sequential_activity_data = {
    "env": my_env,
    "name": "Single run process_1",
    "registry": registry,
    "sub_processes": single_run,
    "postpone_start": True,
}
activity = model.SequentialActivity(**sequential_activity_data)

while_activity = model.WhileActivity(**{
    "env": my_env,
    "name": "while_1",
    "registry": registry,
    "sub_process": activity,
    "condition_event": [{"type":"container", "concept": from_site, "state":"empty"}],
    "postpone_start": False,
})
```

```python
requested_resources1 = {}


single_run = [
    model.MoveActivity(**{
        "env": my_env,
        "name": "sailing empty 2",
        "registry": registry,
        "mover": hopper2,
        "destination": from_site_2,
        "postpone_start": True,
    }),
    model.ShiftAmountActivity(**{
        "env": my_env,
        "name": "Transfer MP 2",
        "registry": registry,
        "processor": hopper2,
        "origin": from_site_2,
        "destination": hopper2,
        "amount": 4,
        "duration": 3600,
        "postpone_start": True,
        "requested_resources": {},
    }),
    model.MoveActivity(**{
        "env": my_env,
        "name": "sailing filled 2",
        "registry": registry,
        "mover": hopper2,
        "destination": to_site,
        "postpone_start": True,
    }),
    model.ShiftAmountActivity(**{
        "env": my_env,
        "name": "Transfer TP 2",
        "registry": registry,
        "processor": hopper2,
        "origin": hopper2,
        "destination": to_site,
        "amount": 4,
        "duration": 3600,
        "postpone_start": True,
        "requested_resources": {},
    })
]


sequential_activity_data = {
    "env": my_env,
    "name": "Single run process_2",
    "registry": registry,
    "sub_processes": single_run,
    "postpone_start": True,
}
activity = model.SequentialActivity(**sequential_activity_data)

while_activity = model.WhileActivity(**{
    "env": my_env,
    "name": "while_2",
    "registry": registry,
    "sub_process": activity,
    "condition_event": [{"type":"container", "concept": from_site_2, "state":"empty"}],
    "postpone_start": False,
})
```

```python
requested_resources1 = {}


single_run = [
    model.MoveActivity(**{
        "env": my_env,
        "name": "sailing empty 3",
        "registry": registry,
        "mover": hopper,
        "destination": from_site_3,
        "postpone_start": True,
    }),
    model.ShiftAmountActivity(**{
        "env": my_env,
        "name": "Transfer MP 3",
        "registry": registry,
        "processor": hopper,
        "origin": from_site_3,
        "destination": hopper,
        "amount": 4,
        "duration": 3600,
        "postpone_start": True,
        "requested_resources": {},
    }),
    model.MoveActivity(**{
        "env": my_env,
        "name": "sailing filled 3",
        "registry": registry,
        "mover": hopper,
        "destination": to_site,
        "postpone_start": True,
    }),
    model.ShiftAmountActivity(**{
        "env": my_env,
        "name": "Transfer TP 3",
        "registry": registry,
        "processor": hopper,
        "origin": hopper,
        "destination": to_site,
        "amount": 4,
        "duration": 3600,
        "postpone_start": True,
        "requested_resources": {},
    })
]


sequential_activity_data = {
    "env": my_env,
    "name": "Single run process_3",
    "registry": registry,
    "sub_processes": single_run,
    "postpone_start": True,
}
activity = model.SequentialActivity(**sequential_activity_data)

while_activity = model.WhileActivity(**{
    "env": my_env,
    "name": "while_3",
    "registry": registry,
    "sub_process": activity,
    "start_event": [
        {
            "and": [
                {"type":"activity", "state":"done", "name": "while_1"},
                {"type":"activity", "state":"done", "name": "while_2"}
            ]
        }
    ],
    "condition_event": [{"type":"container", "concept": from_site_3, "state":"empty"}],
    "postpone_start": False,
})
```

```python
my_env.run()
```
