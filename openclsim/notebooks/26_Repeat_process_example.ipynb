```python
# setup environment
simulation_start = 0
my_env = simpy.Environment(initial_time=simulation_start)
registry = {}
```

```python
basic_activity_data = {
    "env": my_env,
    "name": "Basic activity",
    "registry": registry,
    "duration": 14,
    "postpone_start": True,
}
activity = model.BasicActivity(**basic_activity_data)

repeat_data = {
    "env": my_env,  # The simpy environment defined in the first cel
    "name": "repeat",  # We are moving soil
    "ID": "6dbbbdf7-4589-11e9-bf3b-b469212bff5g",  # For logging purposes
    "registry": registry,
    "sub_process": activity,
    "repetitions": 2,
    "postpone_start": True,
}
repeat_activity = model.RepeatActivity(**repeat_data)


repeat_data_2 = {
    "env": my_env,  # The simpy environment defined in the first cel
    "name": "repeat_2",  # We are moving soil
    "ID": "6dbbbdf7-4589-11e9-bf3b-b469212bff5g",  # For logging purposes
    "registry": registry,
    "sub_process": activity,
    "repetitions": 3,
    "postpone_start": True,
}
repeat_activity_2 = model.RepeatActivity(**repeat_data_2)

sequential_activity_data = {
    "env": my_env,
    "name": "Sequential process",
    "registry": registry,
    "sub_processes": [repeat_activity, repeat_activity_2],
    "postpone_start": False,
}
sequential_activity = model.SequentialActivity(**sequential_activity_data)
```

```python
my_env.run()
```

```python
log_df = pd.DataFrame(sequential_activity.log)
data =log_df[['Message', 'ActivityState', 'Timestamp', 'Value', 'ActivityID']]
data
```
