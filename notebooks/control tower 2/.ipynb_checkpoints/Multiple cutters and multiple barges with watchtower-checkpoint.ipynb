```python
import datetime, time
import simpy

import shapely.geometry
import pandas as pd

import openclsim.core as core
import openclsim.model as model
import openclsim.plot as plot
import numpy as np

from run import run
from IPython.display import clear_output

import matplotlib.pyplot as plt

%matplotlib inline

def extend_id_map(activity, id_map):
    sub_processes=[]
    if hasattr(activity, "sub_processes"):
        sub_processes = activity.sub_processes
    if hasattr(activity, "alternatives"):
        sub_processes = activity.alternatives
        
    for sub_process in sub_processes:
        id_map = extend_id_map(sub_process, id_map)
    
    return {**id_map, activity.id: activity }
```

```python
NR_BARGES=14
NR_CUTTERS=4
TOTAL_AMOUNT=1000
N=10

PRODUCTIONS={
    'barges': [
        {'name': 'barge0', 'unloading_rate': 7, 'v_full': 6, 'capacity': 5},
        {'name': 'barge1', 'unloading_rate': 16, 'v_full': 6, 'capacity': 1},
        {'name': 'barge2', 'unloading_rate': 15, 'v_full': 6, 'capacity': 2},
        {'name': 'barge3', 'unloading_rate': 17, 'v_full': 9, 'capacity': 3},
        {'name': 'barge4', 'unloading_rate': 17, 'v_full': 6, 'capacity': 4},
        {'name': 'barge5', 'unloading_rate': 19, 'v_full': 9, 'capacity': 5},
        {'name': 'barge6', 'unloading_rate': 13, 'v_full': 6, 'capacity': 6},
        {'name': 'barge7', 'unloading_rate': 15, 'v_full': 8, 'capacity': 7},
        {'name': 'barge8', 'unloading_rate': 19, 'v_full': 6, 'capacity': 8},
        {'name': 'barge9', 'unloading_rate': 19, 'v_full': 6, 'capacity': 9},
        {'name': 'barge10', 'unloading_rate': 14, 'v_full': 7, 'capacity': 10},
        {'name': 'barge11', 'unloading_rate': 18, 'v_full': 7, 'capacity': 9},
        {'name': 'barge12', 'unloading_rate': 13, 'v_full': 9, 'capacity': 3},
        {'name': 'barge13', 'unloading_rate': 14, 'v_full': 8, 'capacity': 5},
    ],
    'cutters': [
        {'name': 'cutter0', 'loading_rate': 2},
        {'name': 'cutter1', 'loading_rate': 3},
        {'name': 'cutter2', 'loading_rate': 2},
        {'name': 'cutter3', 'loading_rate': 1},
    ]
}
```

```python
%%time

res = run(
    NR_BARGES=NR_BARGES,
    NR_CUTTERS=NR_CUTTERS,
    TOTAL_AMOUNT=TOTAL_AMOUNT,
    PRODUCTIONS=PRODUCTIONS,
    STRATEGY='optimized',
)
barges = res['barges']
activities = res['activities']
cutters = res['cutters']
to_site = res['to_site']
from_site = res['from_site']
my_env = res['my_env']
input_data = res['input_data']

input_data
```

```python
data = []
for i in range(len(barges.values()) - 1):
    vessel = barges[f'barge{i}']
    activity = activities[f'activity{i}']
    sub_act = activity.sub_processes[0]
    log = pd.DataFrame(vessel.log)
    nr_trips = len(log[ (log.ActivityState == "START")])
    data.append({
        'name':vessel.name, 
        'trips':nr_trips, 
        'production':(nr_trips * vessel.container.get_capacity()) / ((log.Timestamp.max() - log.Timestamp.min()).total_seconds()),
    })
df = pd.DataFrame(data).sort_values(by=['production'],ascending=False)
df
```

```python
data = []
for vessel in [*barges.values(), *cutters.values()]:
    df = (
        pd.DataFrame(vessel.log)
        .replace(id_map)
        .filter(items=["ActivityID", "Timestamp", "ActivityState"])
        .rename(columns={"ActivityID": "Activity"})
    )
    total_time = (df.Timestamp.max() - df.Timestamp.min()).seconds

    df_starts = (
        df[df.ActivityState == "START"]
        .filter(
            ["Activity", "Timestamp"]
        )
        .rename(columns={"Timestamp": "timestamp_start"})
        .set_index("Activity")
    )
    df_stops = (
        df[df.ActivityState == "STOP"]
        .filter(["Activity", "Timestamp"])
        .rename(columns={"Timestamp": "timestamp_stop"})
        .set_index("Activity")
    )
    df2 = pd.concat([df_starts, df_stops], axis=1).reset_index()
    df2['Duration'] = (df2.timestamp_stop - df2.timestamp_start).astype(np.int64)/ 1e9
    active_time = df2.Duration.sum()
    utilisation = 100 * (active_time / total_time)
    data.append({
        "vessel":vessel.name,
        "utilisation": utilisation,
    })
    
pd.DataFrame(data).sort_values(by=['utilisation'],ascending=False)
```

```python
%%time
# assert 1 == 2
data = {}

strategies = ["random", "equal", "optimized"]
for strategy in strategies:
    for i in range(N):
        clear_output(wait=True)
        print(f"{strategy} {i}")
        try:
            res = run(
                NR_BARGES=NR_BARGES,
                NR_CUTTERS=NR_CUTTERS,
                TOTAL_AMOUNT=TOTAL_AMOUNT,
                STRATEGY=strategy,
                PRODUCTIONS=PRODUCTIONS,
            )
            barges = res['barges']
            activities = res['activities']
            cutters = res['cutters']
            to_site = res['to_site']
            from_site = res['from_site']
            my_env = res['my_env']

            activity_map = {}
            for activity in activities.values():
                activity_map = extend_id_map(activity, activity_map)

            id_map = {key:val.name for (key,val) in activity_map.items()}


            for vessel in [*barges.values(), *cutters.values()]:
                df = (
                    pd.DataFrame(vessel.log)
                    .replace(id_map)
                    .filter(items=["ActivityID", "Timestamp", "ActivityState"])
                    .rename(columns={"ActivityID": "Activity"})
                )
                total_time = (df.Timestamp.max() - df.Timestamp.min()).seconds

                df_starts = (
                    df[df.ActivityState == "START"]
                    .filter(
                        ["Activity", "Timestamp"]
                    )
                    .rename(columns={"Timestamp": "timestamp_start"})
                    .set_index("Activity")
                )
                df_stops = (
                    df[df.ActivityState == "STOP"]
                    .filter(["Activity", "Timestamp"])
                    .rename(columns={"Timestamp": "timestamp_stop"})
                    .set_index("Activity")
                )
                df2 = pd.concat([df_starts, df_stops], axis=1).reset_index()
                df2['Duration'] = (df2.timestamp_stop - df2.timestamp_start).astype(np.int64)/ 1e9
                active_time = df2.Duration.sum()
                utilisation = 100 * (active_time / total_time)
                data.setdefault(strategy,{}).setdefault(vessel.name,[]).append(utilisation)
        except Exception as e:
            print(e)

clear_output(wait=True)            
df = []
for stat, dat in data.items():
    df.append({
        'avg_cutter': np.mean([v for (key, val) in dat.items() for v in val if 'cutter' in key]),
        'avg_barge': np.mean([v for (key, val) in dat.items() for v in val if 'barge' in key]),
        **{key: np.mean(val) for (key, val) in dat.items()}
    })
    
display(pd.DataFrame(df, index = data.keys()).T.round())


fig, axs = plt.subplots(2,len(data.keys()),figsize=(15, 15))
i =0
for strat, dat in data.items():
    for j, vessel_type in enumerate(['cutter', 'barge']):
        d2 = {
            'total':[]
        }
        for (key, val) in dat.items():
            if vessel_type in key:
                d2[key] = val
                d2['total'].extend(val)
                
        axs[j][i].set_title(f'{vessel_type} utilisation for the {strat} strategy')
        axs[j][i].boxplot(list(d2.values()))
        axs[j][i].set_ylim([0,105])
        axs[j][i].set_xticklabels(list(d2.keys()), rotation=40)

    
    i+=1
```
